```python
## -----------IMPORT LIBRARIES
import pandas as pd #for dealing with csv import
import os # for joining paths and filenames sensibly
import numpy as np #for the population std
import glob # for finding csv data files
import platform # paths use different dividers on linux vs windows, so we need to test for this
import csv
from pandas.errors import EmptyDataError
```

```python
# square_dot_game_new --> slow
# square_dot_game2_new --> fast
# monkey_game --> monkey
```

```python
root_monkeys= 'C:\\Users\\david\\Downloads\\monkey'
root_slow_SquareDot = 'C:\\Users\\david\\Downloads\\slow'
root_fast_SquareDot2 = 'C:\\Users\\david\\Downloads\\fast'
```

```python
########################## monkey

folders_ = os.listdir(root_monkeys)
all_monkey_files = []

for idx,fold in enumerate(folders_):
    files_ = os.path.join(root_monkeys, folders_[idx])
    search_string=os.path.join(files_,'*.csv') 
    files = glob.glob(search_string)
    all_monkey_files.append(files)
    

monkey_files = [item for sublist in all_monkey_files for item in sublist]


########################## slow

folders_ = os.listdir(root_slow_SquareDot)
all_slow_files = []

for idx,fold in enumerate(folders_):
    files_ = os.path.join(root_slow_SquareDot, folders_[idx])
    search_string=os.path.join(files_,'*.csv') 
    files = glob.glob(search_string)
    all_slow_files.append(files)
    

slow_files = [item for sublist in all_slow_files for item in sublist]


############################## fast


folders_ = os.listdir(root_fast_SquareDot2)
all_fast_files = []

for idx,fold in enumerate(folders_):
    files_ = os.path.join(root_fast_SquareDot2, folders_[idx])
    search_string=os.path.join(files_,'*.csv') 
    files = glob.glob(search_string)
    all_fast_files.append(files)
    

fast_files = [item for sublist in all_fast_files for item in sublist]




summary = []
```

##### Monkey

```python

frames = []
for filename in monkey_files:
    print(filename.split('\\')[-1])
    #import into python using pandas
    try:
        df = pd.read_csv(filename)
    except EmptyDataError:
        print('not enough data')
        print('')
        continue
    
    try:        
        df_sub = df[['Participant','mean_speed', 'max_speed', 'number_trials', 'number_trials_correct', 'valoracion_resp.keys']].iloc[-1]
        df_subset = pd.DataFrame(df_sub).transpose()
        frames.append(df_subset)
        print('ok')
        print('')
    
    except KeyError:
        print('KeyError - not stored')
        print('')
        continue
    
#######################################################################################################

df_ = pd.concat(frames)
df_ ['game'] = 'monkey'
df_=df_.rename(columns = {'valoracion_resp.keys':'valoracion'})

summary.append(df_)
        
```

##### Slow

```python

frames = []
for filename in slow_files:
    print(filename.split('\\')[-1])
    #import into python using pandas
    try:
        df = pd.read_csv(filename)
    except EmptyDataError:
        print('not enough data')
        print('')
        continue
    
    try:        
        df_sub = df[['Participant','mean_speed', 'max_speed', 'number_trials', 'number_trials_correct', 'key_resp_2.keys']].iloc[-1]
        df_subset = pd.DataFrame(df_sub).transpose()
        frames.append(df_subset)
        print('ok')
        print('')
    
    except KeyError:
        print('KeyError - not stored')
        print('')
        continue
            

        
    
#######################################################################################################

df_ = pd.concat(frames)
df_ ['game'] = 'slow'
df_=df_.rename(columns = {'key_resp_2.keys':'valoracion'})
summary.append(df_)
        
```

##### Fast

```python
frames = []
for filename in fast_files:
    print(filename.split('\\')[-1])
    #import into python using pandas
    try:
        df = pd.read_csv(filename)
    except EmptyDataError:
        print('not enough data')
        print('')
        continue
    
    try:        
        df_sub = df[['Participant','mean_speed', 'max_speed', 'number_trials', 'number_trials_correct', 'key_resp_2.keys']].iloc[-1]
        df_subset = pd.DataFrame(df_sub).transpose()
        frames.append(df_subset)
        print('ok')
        print('')
    
    except KeyError:
        print('KeyError - not stored')
        print('')
        continue
            

        
    
#######################################################################################################

df_ = pd.concat(frames)
df_ ['game'] = 'fast'
df_=df_.rename(columns = {'key_resp_2.keys':'valoracion'})
summary.append(df_)
```
