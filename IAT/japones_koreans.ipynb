```python
k_koreano_bueno1 = 500
k_japones_bueno1= 700
k_koreano_bueno2 = 550
k_japones_bueno2= 650

###

j_koreano_bueno1 = 700
j_japones_bueno1= 610
j_koreano_bueno2 = 800
j_japones_bueno2= 550
```

```python
c2 = 'grey'
c1 = 'darkred'
```

```python
plt.figure()
plt.bar(0,k_koreano_bueno1, color=c1, label='coreano-bueno')
plt.bar(1,k_japones_bueno1, color=c2, label='japonés-bueno')

plt.bar(3,k_japones_bueno2, color=c2)
plt.bar(4,k_koreano_bueno2, color=c1)

plt.xticks([0.5, 3.5], ['orden 1', 'orden 2'], fontsize=20)

plt.gca().spines['right'].set_visible(False)  # aesthetics                                                                              # remove right spines
plt.gca().spines['top'].set_visible(False)                                                                                  # remove top spines
plt.gca().get_xaxis().tick_bottom()                                                                                         
plt.gca().get_yaxis().tick_left()
plt.gca().tick_params(direction='in') #direction

plt.ylim(0, 900)
plt.yticks([0,300,600,900], fontsize=14)

plt.ylabel('tiempo reacción (ms)', fontsize=20)

plt.legend(loc=4, fontsize=14)

plt.title('coreanos')
```
