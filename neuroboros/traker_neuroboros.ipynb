```python
def function_b(general, dia):
    g = 0.8*general + 0.2*dia
    return g


def reto_completado(general, reto):
    ## reto puede ser 0 o 1 (no completado (0) o completado (1))
    g = general + 0.05*reto
    if g>1:
        g = 1
    return g
```

```python
days = [np.round(np.mean(np.random.choice([0,0.33, 0.66,1], 5)), 3) for i in range(100)]
retos = [np.random.choice([0,1]) for i in range(100)]

# window_size=5
# numbers_series = pd.Series(days)
# windows = numbers_series.rolling(window_size)
# moving_averages_list = windows.mean()
# moving_averages_list = moving_averages_list.tolist()

weighted_average_b1 = [] ##sin reto
weighted_average = [] ##con reto

b=days[0]
b_r=days[0]

for idx,d in enumerate(days):
    b = function_b(b, d)
    weighted_average_b1.append(b) ##sin reto
    
    b_r = function_b(b_r, d)
    b_r = reto_completado(b_r, retos[idx])
    weighted_average.append(b_r) ##con reto


plt.plot(range(len(weighted_average)), weighted_average, color='darkorange', linestyle='-', linewidth=5, label='weighted + reto')
plt.plot(range(len(weighted_average_b1)), weighted_average_b1, color='darkblue', linestyle='-', label='weighted')
plt.plot(range(len(days)), days, 'r-', linewidth=0.2, label='raw')
retos_y=np.array(retos)-0.025
plt.plot(range(len(retos_y)), retos_y, '*', markersize=3, color= 'k')

plt.legend(fontsize=12, loc = 4, frameon=False)
plt.gca().spines['right'].set_visible(False)  # aesthetics                                                                              # remove right spines
plt.gca().spines['top'].set_visible(False)                                                                                  # remove top spines
plt.gca().get_xaxis().tick_bottom()                                                                                         
plt.gca().get_yaxis().tick_left()
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(0, 1)
plt.xlabel('days', fontsize=14)
plt.ylabel('neuron score', fontsize=14)
```
